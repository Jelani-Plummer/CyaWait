```python
# Group infrequent categories into "Other"
def bucket_infrequent(column, threshold=2):
    # Keep only categories with frequency above threshold
    freq = df[column].value_counts()
    df[column] = df[column].apply(lambda x: x if freq[x] > threshold else 'Other')

# Apply bucketing to the desired columns
bucket_infrequent('Restaurant type')
bucket_infrequent('Food type')

# Now apply one-hot encoding after bucketing
df = pd.get_dummies(df, columns=['Restaurant type', 'Food type'])

df.head()
```
